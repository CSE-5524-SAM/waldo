```python
from ultralytics import YOLO

model = YOLO('yolov8n.yaml').load('yolov8n.pt')  # build from YAML and transfer weights
```

```python
# Train the model
model.train(data='datasets/waldo_data_1/data.yaml', epochs=100, imgsz=640)
model.train(data='datasets/waldo_data_2/data.yaml', epochs=100, imgsz=640)
model.train(data='datasets/waldo_data_3/data.yaml', epochs=100, imgsz=640)
model.export()
```

```python
model = YOLO('runs/detect/train922/weights/best.pt')
```

```python
file_name = 'CropFlou23_jpg.rf.336a1a26b0a8289ce72432f4fe37f187.jpg'
test_path = f'datasets/waldo_data_1/test/images/{file_name}'
pred_result = model(test_path)
bbox = []
for r in pred_result:
    print(r.boxes.xyxy[0])
    bbox = r.boxes.xyxy[0]
```

```python
import cv2
image = cv2.imread(test_path)
#cv2.rectangle(image, (int(bbox[0]), int(bbox[1])), (int(bbox[2]), int(bbox[3])), (0,255,0), 2)
blur = cv2.GaussianBlur(image, (31,31), 0)
min_x = int(bbox[0])
min_y = int(bbox[1])
max_x = int(bbox[2])
max_y = int(bbox[3])
xl = (max_x - min_x) // 2
yl = (max_y - min_y) // 2
min_x = min_x - xl
min_y = min_y - yl
max_x = max_x + xl
max_y = max_y + yl

blur[min_y:max_y, min_x:max_x] = image[min_y:max_y, min_x:max_x]
cv2.rectangle(blur, (min_x, min_y), (max_x, max_y), (0,255,0), 2)
cv2.imwrite('result.jpg', blur)
```
